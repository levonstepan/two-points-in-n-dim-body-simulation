```python
import matplotlib.pyplot as plt
from simulation import simulate, cumulative_function_graph
from ball import Ball, expected_cumulative_function
```

```python
rounds = 10 ** 4
radius = 1
```

```python
body = Ball(radius)
```

```python
simulation = simulate(body, rounds)
```

```python
x_points, y_points = cumulative_function_graph(body, simulation)
expected_y_points = [expected_cumulative_function(x, radius) for x in x_points]
```

```python
plt.plot(x_points, y_points, x_points, expected_y_points)
```

```python
plt.plot(x_points, [y_1 - y_2 for y_1, y_2 in zip(y_points, expected_y_points)])
```
