```python
import matplotlib.pyplot as plt
from simulation import simulate, cumulative_function_graph
from ball import Ball, expected_cumulative_function
```

```python
radius = 1
```

```python
body = Ball(radius)
```

```python
rounds = [10 ** 2, 10 ** 4, 10 ** 6]
```

```python
diffs = []
for r in rounds:
    simulation = simulate(body, r)
    x_points, y_points = cumulative_function_graph(body, simulation)
    expected_y_points = [expected_cumulative_function(x, radius) for x in x_points]
    diff = [y_1 - y_2 for y_1, y_2 in zip(y_points, expected_y_points)]
    diffs.append(diff)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_points, diffs[0], x_points, diffs[1], x_points, diffs[2])
plt.legend(('$10^2$ iterations', '$10^4$ iterations', '$10^6$ iterations'), loc='lower right')
plt.savefig('ball simulation comparison.png')
```
